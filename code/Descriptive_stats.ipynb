```python
import pandas as pd
import glob
import re
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from tqdm import tqdm
%matplotlib inline
```

```python
plt.rc('font', family='sans-serif')
plt.rc('text', usetex=False)
```

```python
def smooth(l, n = 5):
    """moving average filter with window size n"""
    sigma = 0
    res = list( 0 for x in l)
    for i in range(0 , n):
        sigma = sigma + l[i]
        res[i] = sigma / (i + 1)
    for i in range( n, len(l) ):
        sigma = sigma - l[i - n] + l[i]
        res[i] = sigma / n
    return res
```

```python
def unique(text):
    words = text.split(" ")
    n_words = len(words)

    # count each word
    word_count = collections.Counter(words)

    # get the sorted list of unique words
    unique_words = list(word_count.keys())
    unique_words.sort()

    n_unique = len(unique_words)
    return n_unique
```

```python
df = pd.read_pickle('../data/TVG/tvg.pkl')
df['len'] = df['text'].str.split().apply(len)
```

```python
years = df.groupby('year')['len'].count().index.values
word_counts = df.groupby('year')['len'].sum().values
doc_counts = df.groupby('year')['len'].count().values
```

```python
fig, ax = plt.subplots(1,2, figsize=(9,3), dpi=300)

for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(8))
    #axi.yaxis.set_major_locator(plt.MaxNLocator(3))

ax[0].scatter(years, word_counts, facecolors='none', edgecolors='r')
ax[0].plot(years, smooth(word_counts, 5), label='smoothing window=5')
ax[0].set_title('Yearly Word Counts')

ax[1].scatter(years, doc_counts, facecolors='none', edgecolors='r')
ax[1].plot(years, smooth(doc_counts, 5), label='smoothing window=5')
ax[1].set_title('Yearly Document Counts')

ax[0].legend()
ax[1].legend()
plt.savefig('../figures/17c_stats.png')
```
